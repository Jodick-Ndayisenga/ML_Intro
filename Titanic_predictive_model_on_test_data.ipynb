1. **Impute missing cabins based on passenger class:**

Since there is a clear relationship between cabin class and cabin, one approach is to impute missing cabins based on passenger class. For example, passengers in first class (Pclass = 1) are more likely to be in cabins A, B, or C, while passengers in third class (Pclass = 3) are more likely to be in cabins D, E, or F.

```python
import random

# Define a function to impute cabin based on passenger class

class1 = ["A", "B"]
class2 = ["C", "D"]
class3 = ["E", "F", "G"]


def impute_cabin(df):
  for i in range(len(df)):
    if pd.isna(df.loc[i, 'Cabin']):
      if df.loc[i, 'Pclass'] == 1:
        df.loc[i, 'Cabin'] = random.choice(class1)
      elif df.loc[i, 'Pclass'] == 2:
        df.loc[i, 'Cabin'] = random.choice(class2)
      elif df.loc[i, 'Pclass'] == 3:
        df.loc[i, 'Cabin'] = random.choice(class3)
      else:
        df.loc[i, 'Cabin'] = 'C'
  return df

data = impute_cabin(data)
```

```python
data.head(20)
```

## Grabbing the letter from Cabin alone

```python
def shorten_cabin(df):
  for i in range(len(df)):
    if len(df.loc[i, 'Cabin']) > 1:
      df.loc[i, 'Cabin'] = df.loc[i, 'Cabin'][0]
  return df

data = shorten_cabin(data)
data.head(20)
```

```python
#show unique values of cabinet

data['Cabin'].unique()
```

```python
#Check in the whole dataset if there are some missing values

data.isnull().sum()
```

Now the Cabin feature has no missing values but for ages we have to impute the average for each class and replace the missing values with the average of that class

```python
def impute_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    if pd.isnull(Age):
        if Pclass == 1:
            return 37
        elif Pclass == 2:
            return 29
        else:
            return 24

    else:
        return Age
```

```python
data["Age"] = data[["Age", "Pclass"]].apply(impute_age, axis = 1)
```

```python
data.dropna(inplace = True)
```

## Data encoding, now because some of our data are not numerical, we need to encode them for the computer to deal only with numbers. We will use one of the cool functions from sklearn called LabelEncoder to encode our Cabinet letter and for embarked and Sex we will use get dummies

```python
# Encode embarked so that we numerical values for our model

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
sex = pd.get_dummies(data["Sex"], drop_first = True)
embark = pd.get_dummies(data["Embarked"], drop_first = True)
```

Step 5: Split the Data

Separate the features (X) and the target variable (y), then split the data into training and testing sets. We will use one of the commonly used function from sklearn called train_test_split

6. Import of our model

## 7. Training our model
